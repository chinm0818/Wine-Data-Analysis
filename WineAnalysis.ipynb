```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn import tree
import graphviz
```

```python
df = pd.read_csv('WineQuality.csv')
```

## Exploration

```python
df.iloc[:10] #what units are these in? 
# pH, density, alcohol 
# acidity maybe in grams / liter
# assuming all others are also g/liter
```

```python
df.corr()
```

```python
target = df.quality
```

```python
data = df.drop('quality', axis = 1)
```

```python
# lets use this data set to play with decision trees. First. let's do a test train split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(
data, target, test_size = .33, random_state = 42)
```

Playing with decision trees. With and without PCA 

```python
tree1 = tree.DecisionTreeClassifier(max_depth = 3)
tree1 = tree1.fit(X_train, y_train)
```

```python
tree1_predict = tree1.predict(X_test)

#cool. We have a array of predicted ratings. First how where these decided?
```

```python
dot_data = tree.export_graphviz(tree1, out_file = None,
                               feature_names = data.columns,
                               class_names = str([3, 4, 5, 6, 7, 8]),
                               filled = True, rounded = True,
                               special_characters = True)
graph = graphviz.Source(dot_data)
graph.render('wineTree')
```
